# Создание датасета по клиентам

## План действий

Для того, чтобы составить датасет по клиентам нам необходимо агрегировать значения сета по транзакциям. 

Способ этого агрегирования будет напрямую влиять на то, насколько полезным будут собранные в сете по клиентам фичи.

Агрегирование сета до клиентоа будем вести анализируя каждую из фичей сета транзакций. Анализ должен давать ответ на вопрос: какой из способов агрегирования будет наиболее полезным для определения гендера.

## Первые гипотезы


- **Количественные показатели** (суммы оплат) считаем в такие агрегаты как среднее, минимальное и максимальное значение. Количество повторяющихся операций.

- **Со справочными показателями** (категория операций или тип операции) интереснее. По ним будем определять, насколько каждая из них "мужская" или "женская". Агрегатом в таком случае будет являться доля транзакций каждого гендера.

- **Показатели даты и времени** попробуем сгруппировать по наиболее частым времени / дате недели совершения операций. Тут будет попытка отделить мужчин и женщин по среднему времени совершения операций.

Таким образом мы попробуем собрать фичи **клиентов**, которые могут указывать на их принадлежность какому-то из гендеров.
